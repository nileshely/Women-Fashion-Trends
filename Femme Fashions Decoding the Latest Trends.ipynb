```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/dataset/summer-products.csv")
```

```python
df.head()
```

```python
# Set pink color palette
sns.set_palette(sns.color_palette(["#FFA7A7", "#FF7F7F", "#FF5252", "#FF1A1A", "#FF0000"]))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['units_sold'], bins=30, kde=True)
plt.title('Distribution of Units Sold', fontsize=16)
plt.xlabel('Units Sold', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='rating', data=df, color='pink')
plt.title('Price vs Rating', fontsize=16)
plt.xlabel('Price', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='rating', data=df, color='pink')
plt.title('Distribution of Ratings', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['rating_count'], bins=30, kde=True, color='pink')
plt.title('Distribution of Rating Counts', fontsize=16)
plt.xlabel('Rating Count', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['retail_price'], bins=30, kde=True, color='pink')
plt.title('Distribution of Retail Price', fontsize=16)
plt.xlabel('Retail Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rating_count', y='units_sold', data=df, color='pink')
plt.title('Rating Count vs Units Sold', fontsize=16)
plt.xlabel('Rating Count', fontsize=14)
plt.ylabel('Units Sold', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['shipping_option_price'], bins=30, kde=True, color='pink')
plt.title('Distribution of Shipping Option Price', fontsize=16)
plt.xlabel('Shipping Option Price', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['merchant_rating'], bins=30, kde=True, color='pink')
plt.title('Distribution of Merchant Ratings', fontsize=16)
plt.xlabel('Merchant Rating', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['inventory_total'], bins=30, kde=True, color='pink')
plt.title('Distribution of Inventory Total', fontsize=16)
plt.xlabel('Inventory Total', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='retail_price', y='units_sold', data=df, color='pink')
plt.title('Retail Price vs Units Sold', fontsize=16)
plt.xlabel('Retail Price', fontsize=14)
plt.ylabel('Units Sold', fontsize=14)
plt.show()
```
